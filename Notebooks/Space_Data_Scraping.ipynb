```python
import pandas as pd
import requests
from bs4 import BeautifulSoup 
```

```python
url = "https://en.wikipedia.org/wiki/2020_in_spaceflight"
table_class = "wikitable collapsible mw-collapsible mw-made-collapsible"

# GET request
response = requests.get(url)
print(response.status_code)
```

```python
# parse data from the html into a beautifulsoup object
soup = BeautifulSoup(response.text, 'html.parser')
Spaceflight_2020 = soup.find('table',{'class':"wikitable collapsible"})
print(Spaceflight_2020)
```
